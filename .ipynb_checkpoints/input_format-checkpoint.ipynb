```python
import pandas as pd
import json
from datetime import datetime
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

# Load fires csv

```python
df = pd.read_csv('./Documents/2020_with_emissions.csv')
pd.set_option('display.max_columns', None)
print(df.head())
```

```python
sorted_df = df.sort_values(by = 'SumOfPM25F_Lb', ascending=False)
sorted_df['cum_PM2.5'] = sorted_df['SumOfPM25F_Lb'].cumsum()/sorted_df['SumOfPM25F_Lb'].sum()
print(sorted_df['cum_PM2.5'][0:15])
plt.plot(list(sorted_df['cum_PM2.5'][0:50]))
plt.xlabel('2020 Fires sorted descending by PM2.5 emissions')
plt.ylabel('2020 cumulative PM2.5 emissions(%)')
```

# Load all fires

```python
data = {}
data['fires'] = []
for i in tqdm((range(len(df)))):
    temp_fire = {}
    temp_fire['event_of'] = {}
    temp_fire['name'] = df['FIRE_NAME'][i]
    temp_fire['activity'] = {}
    temp_fire['activity']['name'] = "test"
    temp_fire['activity']['active_areas'] = []
    start_date = df['ALARM_DATE_MAIN'][i]
    end_date = df['CONT_DATE'][i]

    temp_active_area = {}
    temp_active_area['start'] = str(datetime.strptime(start_date, "%d-%b-%y").isoformat(timespec='seconds'))
    try:
        temp_active_area['end'] = str(datetime.strptime(end_date, "%d-%b-%y").isoformat(timespec='seconds'))
    except ValueError:
        continue
    temp_active_area['utc_offset'] = '-08:00'
    temp_active_area['country'] = 'USA'
    temp_active_area['state'] = df['STATE'][i]
    temp_active_area['ecoregion'] = 'western'
    temp_fire['activity']['active_areas'].append(temp_active_area)
    data['fires'].append(temp_fire)
```

```python
# Print in human-readable format
print(json.dumps(data, indent=4))

# Save to .json file
with open('output.json', 'w') as outfile:
    json.dump(data, outfile)
```
